
#### 3. Who has played the most all star games?


```python
pd.read_sql_query('''SELECT playerid, COUNT(gameid) as GameCount
                     FROM allstar
                     GROUP BY playerid
                     ORDER BY GameCount DESC
                     LIMIT 5;''',cnx)
```
