```python
import requests
from bs4 import BeautifulSoup
import re
import pandas as pd
```

### url format for inventory data
https://waterdata.usgs.gov/nwis/inventory/?site_no=07374000&agency_cd=USGS


```python
# url = 'https://waterdata.usgs.gov/nwis/inventory/?site_no=06893000&agency_cd=USGS'
url = 'https://waterdata.usgs.gov/nwis/inventory/?site_no=07374000&agency_cd=USGS'

response = requests.get(url)
response.status_code
```

```python
page = response.text
soup = BeautifulSoup(page,"lxml")
```

```python
#print(soup.prettify())
```

```python
station_name = soup.find('h2').text
print(station_name)
```

```python
description = soup.find(id='stationTable').find_all('dl')[0]
print(description.prettify())
```

```python
drainage_string = description.find_all('dd')[2].text
print(drainage_string)
```

```python
datum_string = description.find_all('dd')[3].text
print(datum_string)
```

```python
inventory = soup.find(id='stationTable').find_all('dl')[1]
print(inventory.prettify())
```

```python
k = 4
print(inventory.find_all('tr')[k].find_all('td')[0].text.strip(),
      inventory.find_all('tr')[k].find_all('td')[1].text.strip(),
      inventory.find_all('tr')[k].find_all('td')[2].text.strip(),
      inventory.find_all('tr')[k].find_all('td')[3].text.strip())
```
