```python
# Generate a confusion matrix plot: 

def plot_confusion_matrix(cm,title='Confusion matrix', cmap=plt.cm.Blues):
    ax=plt.imshow(cm, interpolation='nearest',cmap=cmap)
#     plt.title(title)
    plt.colorbar()
    plt.tight_layout()
#     ax.set_xticklabels(['a','b','c'])
    plt.ylabel('Poisonous',fontsize=12)
    plt.xlabel('Predicted',fontsize=12)
    plt.text(0-cm[0][0]**0.5/500,0,cm[0][0],fontsize=15,fontweight='bold',color='#AA0000')
    plt.text(0-cm[0][1]**0.5/500,1,cm[0][1],fontsize=15,fontweight='bold',color='#AA0000')
    plt.text(1-cm[1][0]**0.5/500,0,cm[1][0],fontsize=15,fontweight='bold',color='#009977')
    plt.text(1-cm[1][1]**0.5/500,1,cm[1][1],fontsize=15,fontweight='bold',color='w')
    
#Could be a typical function for classifying:

def train_score(classifier,x,y,test_size=0.3):
    xtrain, xtest, ytrain, ytest = train_test_split(x, y, stratify=y, test_size=test_size,random_state=1111)
#     ytrain=np.ravel(ytrain)
    clf = classifier.fit(xtrain, ytrain)
    # accuracy for test & train:
    train_acc=clf.score(xtrain, ytrain)
    test_acc=clf.score(xtest,ytest)
    print("Training Data Accuracy: %0.2f" %(train_acc))
    print("Test Data Accuracy:     %0.2f" %(test_acc))
    
    y_true = ytest
    y_pred = clf.predict(xtest)


    conf = confusion_matrix(y_true, y_pred,labels=[1,0])
    print(conf)

    print ('\n')
    print ("Precision:              %0.2f" %(conf[0, 0] / (conf[0, 0] + conf[1, 0])))
    print ("Recall:                 %0.2f"% (conf[0, 0] / (conf[0, 0] + conf[0, 1])))
    
    cm=confusion_matrix(y_true, y_pred, labels=[1,0])
    
    plt.figure()
    plot_confusion_matrix(cm)
    
    return(clf)

def plot_roc(classifier,x,y, test_size=0.3):
    xtrain, xtest, ytrain, ytest = train_test_split(x, y, stratify=y, test_size=test_size, random_state=1111)
    clf = classifier.fit(xtrain, ytrain)
    
    # calculate and print accuracy score
    train_acc=clf.score(xtrain, ytrain)
    test_acc=clf.score(xtest,ytest)
    print("Training Data Accuracy: %0.2f" %(train_acc))
    print("Test Data Accuracy:     %0.2f" %(test_acc))
    print(classification_report(ytest, clf.predict(xtest)))
    
    # calculate ROC
    y_score=clf.predict_proba(xtest)[:,1]
    fpr, tpr,_ = roc_curve(ytest, y_score)
    roc_auc = auc(fpr, tpr)
    
    # calculate confusion matrix
    y_true = ytest
    y_pred = clf.predict(xtest)
    conf = confusion_matrix(y_true, y_pred, labels=[1,0])
#     conf = confusion_matrix(y_true, y_pred)
    print(conf)

    # make plots
    fig,ax = plt.subplots(1,2,figsize=(10,4))
    # plot ROC
    ax = plt.subplot(1,2,1)
    plt.plot([0,1],[0,1],'k:')
    plt.plot(fpr,tpr)
    plt.xlabel('FPR',fontsize=12)
    plt.ylabel('TPR',fontsize=12)
    plt.axis('scaled')
    plt.tight_layout()
    # plot confusion matrix
    ax = plt.subplot(1,2,2)
#     sns.heatmap(conf, annot=True, fmt='d',annot_kws={"size": 20}, 
#                 cmap='Blues', cbar=False,
#                 xticklabels=['pred_edible','pred_poison'],
#                 yticklabels=['true_edible','true_poison'])
    plt.imshow(conf, interpolation='nearest',cmap=plt.cm.bone_r)
#     plt.colorbar()
#     plt.tight_layout()
#     plt.ylabel('True',fontsize=12)
#     plt.xlabel('Predicted',fontsize=12)
    plt.text(0,0,conf[0][0],fontsize=15,fontweight='bold',color='w',ha='center')
    plt.text(1,0,conf[0][1],fontsize=15,fontweight='bold',color='#CC0000',ha='center')
    plt.text(0,1,conf[1][0],fontsize=15,fontweight='bold',color='k',ha='center')
    plt.text(1,1,conf[1][1],fontsize=15,fontweight='bold',color='w',ha='center')
    ax.set_xticks([0,1])
    ax.set_xticklabels(['Poisonous\n(Predicted)','Edible\n(Predicted)'],fontsize=12)
    ax.set_yticks([0,1])
    ax.set_yticklabels(['Poisonous\n(True)','Edible\n(True)'],fontsize=12)
    plt.tight_layout()
    
    return clf
```

### Logistic Regression

```python
clf = LogisticRegression()
plot_roc(clf,X,y,test_size=0.3)
```

### Decision Tree

```python
clf = DecisionTreeClassifier(max_depth=5)
plot_roc(clf,X,y,test_size=0.3)
```

```python
dot_data = StringIO()
export_graphviz(clf, out_file=dot_data,  
                filled=True, rounded=True,
                special_characters=True,feature_names = X.columns)
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
Image(graph.create_png())
```

### Random Forest

```python
rf = RandomForestClassifier(n_estimators=100, max_depth=1)
plot_roc(rf,X,y,test_size=0.3)
```

```python
dot_data = StringIO()
export_graphviz(rf, out_file=dot_data,  
                filled=True, rounded=True,
                special_characters=True,feature_names = X.columns)
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
Image(graph.create_png())
```

### KNN

```python
X_train,X_test,y_train,y_test = train_test_split(X, y, test_size=0.3, random_state=4444)
scores = []
nb_num = range(1,21)
for k in nb_num:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    y_pred = knn.predict(X_test)
    scores.append(accuracy_score(y_test, y_pred))
    print(k,accuracy_score(y_test, y_pred))


fig,ax = plt.subplots(1,1,figsize=(7,4));
ax.plot(nb_num,scores,'b')
ax.set_xticks(nb_num);
ax.grid(color='k', linestyle='--', linewidth=1, alpha=0.5)
ax.set_xlabel('# of neighbors',fontsize=12);
ax.set_ylabel('accuracy',fontsize=12);

knn = KNeighborsClassifier(n_neighbors=10)
train_score(knn,X,y)
```
