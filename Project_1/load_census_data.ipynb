```python
import simplejson as json

file = os.path.join(directory, "NYC_2010_Census_Tracts.geojson")
bx = open(file)
bx = json.loads(bx.read())

code = []
boro = []
pol = []
center = []
for k in range(len(bx['features'])):
    code.append(int(bx['features'][k]['properties']['boro_ct2010']))
    boro.append(bx['features'][k]['properties']['boro_name'])
    pol.append(np.array(bx['features'][k]['geometry']['coordinates'][0][0]))
    center.append(pol[k].mean(axis=0))
```

```python
boro_dict = {'Brooklyn':'3604', 
             'Queens': '3608',
             'Bronx': '3600',
             'Manhattan':'3606',
             'Staten Island':'3608'}
boro_code = [boro_dict[x] for x in boro]
```

```python
ct_code = [boro_code[k]+str(code[k]) for k in range(len(code))]
```

```python
ct = pd.DataFrame({'CensusTract': [int(x) for x in ct_code], 
                   'center_loc': center})
```

```python
def FindLocation(xy_list,df2):
    ind = []
    for i in range(len(xy_list)):
        xy = xy_list[i]
        dif = np.array([np.abs(df2.iloc[k].center_loc[0]-xy[0])+np.abs(df2.iloc[k].center_loc[1]-xy[1]) 
                        for k in range(len(df2))])
        
        ind.append(dif.argmin())

    return(ind)
        
def FindCensusTract(df1,df2):
    df1['CensusTract'] = np.nan
    xy_list = list(df1.coordinates)
    for i in range(len(xy_list)):
        xy = np.array(xy_list[i])
        dif = np.array([np.abs(df2.iloc[k].center_loc[0]-xy[0])+np.abs(df2.iloc[k].center_loc[1]-xy[1]) 
                        for k in range(len(df2))])
        ind = dif.argmin()
        df1.at[i,'CensusTract'] = df2['CensusTract'][ind]

    return(df1)
```
