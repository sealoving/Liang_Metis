```python
pol = [(40.757320, -74.006629),
       (40.754569, -73.995595),
       (40.762716, -73.987214),
       (40.761446, -73.969755),
       (40.705987, -74.011099),
       (40.713081, -74.015429)]
```

```python
from shapely.geometry import Point
from shapely.geometry.polygon import Polygon

point = Point(0.5, 0.5)
polygon = Polygon([(0, 0), (0, 1), (1, 1), (1, 0)])
print(polygon.contains(point))
```

```python
def InTech(latlong,poly):
    pt = Point(latlong)
    pol = Polygon(poly)
    return(pol.contains(pt))
```

```python
df['InTechZone'] = df.geo.apply(lambda x: InTech(x,pol))
```
